```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

# Parameters
prob = scipy.stats.beta.rvs(1, 1)
print(f"p={round(prob,4)}")
num_steps = 10000

# Initialize variables
outcomes = np.array([0, 0])
first_order_total_loss= 0
first_order_mean_losses = []
second_order_total_loss = 0
second_order_mean_losses = []

# Simulation loop
for i in range(num_steps):
    first_order_pred = np.argmax(outcomes)
    second_order_pred = (outcomes+1/2)/(outcomes.sum()+1)
    flip = scipy.stats.bernoulli.rvs(prob)
    outcomes[flip] += 1
    first_order_total_loss += first_order_pred != flip
    first_order_mean_losses.append(first_order_total_loss / (i + 1))
    second_order_total_loss += -np.log2(second_order_pred[flip])
    second_order_mean_losses.append(second_order_total_loss / (i + 1))

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot mean rewards
axs[0].plot(first_order_mean_losses)
axs[0].plot(np.exp(-np.array(second_order_mean_losses)))
axs[0].set_title('Mean Losses')
axs[0].set_xlabel('Steps')
axs[0].set_ylabel('Mean Loss')

# Plot scaled error
first_order_scaled_error = np.sqrt(np.arange(1, num_steps + 1)) * (np.array(first_order_mean_losses) - min(prob, 1-prob))
axs[1].plot(first_order_scaled_error)
second_order_scaled_error = np.sqrt(np.arange(1, num_steps + 1)) * (np.array(second_order_mean_losses)) - (-prob*np.log(prob)-(1-prob)*np.log(1-prob))
axs[1].plot(second_order_scaled_error)
axs[1].set_title('Scaled Error')
axs[1].set_xlabel('Steps')
axs[1].set_ylabel('Error')

# Show plots
plt.tight_layout()
plt.show()
```
